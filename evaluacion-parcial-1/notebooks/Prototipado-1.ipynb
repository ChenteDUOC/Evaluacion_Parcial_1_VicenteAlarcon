```python
# Detectar nulos y duplicados en todos los datasets 
for nombre, df in [("Ventas", df_ventas), ("Productos", df_productos), 
                  ("Clientes", df_clientes), ("Devoluciones", df_devoluciones)]:
    print(f"\nDiagnóstico {nombre}:")
    print(f"- Nulos totales: {df.isnull().sum().sum()}")
    print(f"- Filas duplicadas: {df.duplicated().sum()}")
```

```python
def preprocess_ventas(df: pd.DataFrame) -> pd.DataFrame:
    df_clean = df.copy().drop_duplicates()
    df_clean = df_clean.dropna(subset=['id_venta', 'id_cliente', 'id_producto'])
    
    df_clean['fecha'] = pd.to_datetime(df_clean['fecha'], dayfirst=True, errors='coerce')
    df_clean = df_clean.dropna(subset=['fecha'])
    
    cols_str = ['metodo_pago', 'canal_venta']
    for col in cols_str:
        if col in df_clean.columns:
            df_clean[col] = df_clean[col].astype(str).str.strip().str.title()
            df_clean[col] = df_clean[col].replace('Nan', np.nan)
            df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
            
    cols_num = ['precio_unitario', 'cantidad']
    for col in cols_num:
        if col in df_clean.columns:
            df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
            df_clean[col] = df_clean[col].fillna(df_clean[col].median())
    
    return df_clean
```

```python
def preprocess_productos(df: pd.DataFrame) -> pd.DataFrame:
    df_clean = df.copy().drop_duplicates()
    df_clean = df_clean.dropna(subset=['id_producto']) 
    
    cols_str = ['categoria', 'subcategoria', 'proveedor']
    for col in cols_str:
        if col in df_clean.columns:
            df_clean[col] = df_clean[col].astype(str).str.strip().str.title()
            df_clean[col] = df_clean[col].replace('Nan', np.nan)
            df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
            
    cols_num = ['precio_lista', 'stock']
    for col in cols_num:
        if col in df_clean.columns:
            df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')
            df_clean[col] = df_clean[col].fillna(df_clean[col].median())
            
    return df_clean
```

```python
def preprocess_clientes(df: pd.DataFrame) -> pd.DataFrame:
    df_clean = df.copy().drop_duplicates()
    df_clean = df_clean.dropna(subset=['id_cliente'])
    
    df_clean['fecha_registro'] = pd.to_datetime(df_clean['fecha_registro'], dayfirst=True, errors='coerce')
    df_clean = df_clean.dropna(subset=['fecha_registro'])
    
    cols_str = ['region', 'ciudad', 'segmento']
    for col in cols_str:
        if col in df_clean.columns:
            df_clean[col] = df_clean[col].astype(str).str.strip().str.title()
            df_clean[col] = df_clean[col].replace('Nan', np.nan)
            df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
            
    return df_clean
```

```python
def preprocess_devoluciones(df: pd.DataFrame) -> pd.DataFrame:
    df_clean = df.copy().drop_duplicates()
    df_clean = df_clean.dropna(subset=['id_devolucion', 'id_venta'])
    
    df_clean['fecha_devolucion'] = pd.to_datetime(df_clean['fecha_devolucion'], dayfirst=True, errors='coerce')
    df_clean = df_clean.dropna(subset=['fecha_devolucion'])
    
    cols_str = ['motivo', 'estado']
    for col in cols_str:
        if col in df_clean.columns:
            df_clean[col] = df_clean[col].astype(str).str.strip().str.title()
            df_clean[col] = df_clean[col].replace('Nan', np.nan)
            df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
            
    return df_clean
```

```python
# === EJECUCIÓN Y VERIFICACIÓN DE LIMPIEZA ===

# 1. Ejecutar las 4 funciones
df_ventas_limpio = preprocess_ventas(df_ventas)
df_productos_limpio = preprocess_productos(df_productos)
df_clientes_limpio = preprocess_clientes(df_clientes)
df_devoluciones_limpio = preprocess_devoluciones(df_devoluciones)

# 2. Verificar resultados (Nulos y Duplicados deben ser 0)
datasets_limpios = {
    "Ventas": df_ventas_limpio,
    "Productos": df_productos_limpio,
    "Clientes": df_clientes_limpio,
    "Devoluciones": df_devoluciones_limpio
}

print("=== REPORTE DE VERIFICACIÓN POST-LIMPIEZA ===")
for nombre, df in datasets_limpios.items():
    nulos = df.isnull().sum().sum()
    duplicados = df.duplicated().sum()
    print(f"{nombre}: {nulos} Nulos | {duplicados} Duplicados | Shape: {df.shape}")
```

```python
# Analíticas de Outliers
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=df_ventas_limpio, y='precio_unitario', color='salmon')
plt.title('Outliers en Precio Unitario (Ventas)')

plt.subplot(1, 2, 2)
sns.boxplot(data=df_productos_limpio, y='stock', color='skyblue')
plt.title('Outliers en Stock (Productos)')

plt.tight_layout()
plt.show()
```

```python
# Unimos ventas y productos para ver correlación
df_join_corr = df_ventas_limpio.merge(df_productos_limpio, on='id_producto', how='inner')

# Seleccionamos solo columnas numéricas
cols_num = df_join_corr.select_dtypes(include=[np.number]).columns

# Coeficiente de Pearson (Lineal)
print("--- Correlación de Pearson ---")
pearson_corr = df_join_corr[cols_num].corr(method='pearson')
display(pearson_corr)

# Coeficiente de Spearman (No Lineal / Rangos)
print("\n--- Correlación de Spearman ---")
spearman_corr = df_join_corr[cols_num].corr(method='spearman')
display(spearman_corr)

# Graficar Heatmap de Pearson
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, cmap='RdYlGn', fmt='.2f')
plt.title('Matriz de Correlación de Pearson')
plt.show()
```
